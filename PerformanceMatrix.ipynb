```python
from sklearn.metrics import accuracy_score , precision_score ,recall_score , f1_score , confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

#True ans (actual answers)
y_actual = [1,0,1,1,0,1,0]
# model predictions 
y_predicted = [1,0,1,0,0,1,1]

#evaluation 
print("Accuracy :",round(accuracy_score(y_actual , y_predicted),2)*100,"%")
print("Precision :", round(precision_score(y_actual , y_predicted),2)*100,"%")
print("recall :",round(recall_score(y_actual , y_predicted),2)*100,"%")
print("f1_score:", round(f1_score(y_actual , y_predicted),2)*100,"%")
```

```python
cm = confusion_matrix(y_actual , y_predicted)
print("confusion matrix : \n",cm)
plt.figure(figsize=(8,6))
sns.heatmap(cm , annot=True , fmt='d' ,cmap='Blues')
plt.title('Confusion matrix')
plt.xlabel("predicted values")
plt.ylabel("actual values ")
plt.show()
```
