```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data of cgpa vs lpa")
print(df)
```

```python
X=df.iloc[:,0:1] #cgpa
Y=df.iloc[:,1:2] #Lpa
```

```python
plt.scatter(X,Y)
plt.xlabel("CGPA")
plt.ylabel("LPA")
plt.show()
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

```

```python
#More the data point is closer to the line more accurate is the prediction according to the data (if data is correct)
model = LinearRegression()
model.fit(X,Y)
cgpa = float(input("Enter the cgpa you have got:"))

#model is tested on the random value(cgpa)
y_pred1 = model.predict([[cgpa]])

#model is tested on the same input value (x) 
y_predicted = model.predict(X)
plt.scatter(X , Y)
plt.plot(X,y_predicted ,color='red' ,linewidth=2)
plt.scatter(cgpa , y_pred1 ,color='green',s=100,label="your predicted value ")
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
model.fit(X_train,Y_train)
```

```python

print(model.predict([[8.6]])[0][0]) # tested on the random value
print("Slope (m):", model.coef_)
print("Intercept (c):", model.intercept_)
```

```python
modelY_pred = model.predict(X_test)
print(modelY_pred)
```

```python
print(Y_test) #actual values
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score,root_mean_squared_error
print(("MAE",mean_absolute_error(Y_test , modelY_pred)))
print("MSE ",mean_squared_error(Y_test , modelY_pred))
print("RMSE",root_mean_squared_error(Y_test,modelY_pred))
print("r2_score",r2_score(Y_test , modelY_pred))
R2 = r2_score(Y_test , modelY_pred)
```

```python
# Adjusted R2 sore 
print(X_test.shape) 
#formula
adjustedR2_score= float(1-((1-R2)*(40-1)/(40-1-1)))
print(adjustedR2_score)
```
