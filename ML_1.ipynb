```python
num_cols = ['age','bp','sg','al','su','bgr','bu',
            'sc','sod','pot','hemo','pcv','wc','rc']

notnum_cols = [c for c in df.columns if c not in num_cols]
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```
