## scatter plot (numerical - numerical)

Analysis of multiple columns on a single plot 

```python
sns.scatterplot(x='total_bill',y='tip',data=tips ,hue=tips['sex'],style=tips['smoker'])
plt.show()
```

## Bar plot (numerical - catagorical )

```python
sns.barplot(x='Pclass',y='Fare',data=df,hue=df['Sex']) # X-catagory ,Y-numerical
```

```python
df.groupby(['Pclass', 'Sex']).size()
```

```python
df.groupby(['Pclass', 'Sex'])['Fare'].mean()
```

## 3 Box plot (numerical - catagorical )

```python
sns.boxplot(x='Sex',y='Age',data=df,hue=df['Survived'])
```

## 4 Distplot or Kdeplot (Numerical - catagorical)

```python
sns.kdeplot(df[df['Survived']==0]['Age'])
sns.kdeplot(df[df['Survived']==1]['Age'])
```
