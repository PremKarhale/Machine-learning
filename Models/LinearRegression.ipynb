```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("data of cgpa vs lpa")
print(df)
```

```python
X = df.iloc[:,0:1] #input feature 
Y = df.iloc[:,-1]#output feature 
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X , Y , test_size=0.2 , random_state= 2)
```

```python
model = LinearRegression()
model.fit(X_train , Y_train)
```

```python
print(X_test)
```

```python
print(Y_test)
```

```python
# print(X_test.iloc[0].values[0],"CGPA")
# y_predicted =model.predict(X_test.iloc[0].values.reshape([1,1]))
y_predicted=(model.predict(X_test))
# print(f"This is my model predicted result that on {X_test.iloc[0].values[0]} CGPA i get : ",y_predicted[0],"LPA")
print(y_predicted)
```

```python
R2 = r2_score(Y_test,y_predicted)
print(R2)
```

```python
#predicted the lpa for any new or random cgpa 
print(df.max())
result = model.predict([[ 9.58]])
print(result)
```
