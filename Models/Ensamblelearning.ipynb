```python
X_cat =[]
for col in df.columns:
    if df[col].dtype == 'object':
        X_cat.append(col)

X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oe = OneHotEncoder(sparse_output=False)
X_encoded = oe.fit_transform(df[X_cat])
X_encoded
```

```python
X_encoded = pd.DataFrame(X_encoded , columns=oe.get_feature_names_out(X_cat))
```

```python
X_encoded
```

```python
df=df.drop(columns=X_cat)
```

```python
df = pd.concat([df,X_encoded],axis = 1)
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['depression_label'])
y=df['depression_label']
```

```python
X.shape
```

```python
y
```

```python
X_train , X_test , y_train , y_test =train_test_split(X , y ,test_size=0.2 , random_state=20)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
X_train
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
Accuracy = accuracy_score(y_test,y_pred)
print("the Accuracy is : ", Accuracy)
```

```python
svm = SVC(kernel='poly',degree=2) # polly because data is not continuous 
svm.fit(X_train,y_train)
y_pred = svm.predict(X_test)
Accuracy = accuracy_score(y_test,y_pred)
Accuracy 
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import cross_val_score
```

with cross val score 

```python
estimators = [('lr',LogisticRegression()),('svm',SVC(probability=True)),('knn',KNeighborsClassifier())]
ensembel_model = VotingClassifier(estimators=estimators , voting='soft')
accuracy = cross_val_score(ensembel_model , X , y , cv = 10 , scoring="accuracy")
mean_accuracy = np.round(accuracy.mean(),2)
mean_accuracy
```

without cross val score 

```python
estimators = [('lr',LogisticRegression()),('svm',SVC()),('knn',KNeighborsClassifier())]
ensembel_model = VotingClassifier(estimators=estimators , voting='hard')
ensembel_model.fit(X_train,y_train)
y_pred = ensembel_model.predict(X_test)
ans=accuracy_score(y_test,y_pred)
ans
```
