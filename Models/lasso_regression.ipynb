## Laso Regression 
- it is a regularisation technique.  MSE + (aplha*|m| )
- increasing the value of alpha / lambda decreases the value of m , can also make it zero 

- lasso regression is most preffered regularisation techniqure , because it internally does feature selection and removes the most less important features. by making their weights to Zero

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
X , y = make_regression(n_samples=500 , n_features = 1, n_informative=1,n_targets=1,noise = 30 , random_state=2)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
r2_score(y,lr.predict(X))
```

```python
from sklearn.linear_model import Lasso
```

```python
la = Lasso(alpha=6)
la.fit(X,y)
print("coef :",la.coef_)
print("intercept:",la.intercept_)
```

```python
y_pred = la.predict(X)
print("r2_score :",r2_score(y,y_pred))
```

```python
plt.plot(X,y,'b. ')
plt.plot(X,lr.predict(X),color="red",label="alpha = 0")
plt.plot(X,la.predict(X),color="yellow",label="alpha = 10")
plt.legend()
```
