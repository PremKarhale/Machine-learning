```python
# chi squre test 
from scipy.stats import chi2_contingency
table = pd.crosstab(df['price'],df['facing'])
chi2,p,dof,exp=chi2_contingency(table)
print('chi2',chi2)
print("p-value",np.round(p,2)) # p < 0.05 → Significant association  | p ≥ 0.05 → No statistical relationship
```

```python
#seperated the data into num_col or not_num_col 
num_col = ['price','price_per_sqft','bedRoom','bathroom','floorNum'] 
cat_col =  ['property_type', 'balcony', 'additionalRoom', 'rating', 'facing', 'agePossession']
print(cat_col)
```

```python
print(data[num_col].dtypes)
print(data[cat_col].dtypes)
```

```python
# data.loc[13,'bedRoom']=np.nan  used to insert the NAN values in between the df
print(data[num_col])
print(data[num_col].isnull().sum())

```

```python
data[num_col] = data[num_col].fillna(data[num_col].mean())
```

```python
data[cat_col] = data[cat_col].fillna(data[cat_col].mode().iloc[0])
```

```python
print(data[num_col].isnull().sum())
```

```python
print(data[cat_col].isnull().sum())
```

```python
print(data.isnull().sum().sum())
```

```python
#convert all the catagorical data into numeric data 
data = pd.get_dummies(data , columns=cat_col,drop_first=True)
```

```python
data = data.replace({True:1,False:0})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
print(data.columns)
print(data.dtypes)
data.shape
```

```python
X = data.drop(columns=['price']) # input features 
Y = data['price'] #output features 
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y , test_size=0.2 , random_state=2)
```

```python
model = LinearRegression()
model.fit(X_train , Y_train)
```
