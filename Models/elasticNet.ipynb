```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Linear Regression 
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred1 = lr.predict(X_test)
print("R2 score : ",r2_score(y_test , y_pred1))
```

```python
# Ridge Regression 
reg = Ridge(alpha = 0.2)
reg.fit(X_train,y_train)
y_pred2 = reg.predict(X_test)
print("R2 score : ",r2_score(y_test , y_pred2))
```

```python
# Lasso Regression 
# for lasso aplha must be kept small, it very rapidly makes the value of weights to zero and decreases the slope
lso = Lasso(alpha = 0.01) 
lso.fit(X_train,y_train)
y_pred3 = lso.predict(X_test)
print("R2 score : ",r2_score(y_test , y_pred3))
```

```python
# Elastic net 
elastic = ElasticNet(alpha = 0.005,l1_ratio = 0.9)
elastic.fit(X_train,y_train)
y_pred4 = elastic.predict(X_test)
print("R2 score : ",r2_score(y_test,y_pred4))
```
