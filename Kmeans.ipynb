```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(x='cgpa',y='iq',data=df)
plt.xlabel('CGPA')
plt.ylabel('IQ')
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(df)  # doing both fit and predict together !!
    wcss.append(km.inertia_)
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
X = df.iloc[:,:].values # convert all the datapoints to a numpy array 
km = KMeans(n_clusters=4)
Y_means=km.fit_predict(X)
```

```python
Y_means # divide all the data points into multiple clusters !!
```

```python
X[Y_means==0] #boolean indexing  # data points who are in the first clusters 
```

```python
sns.scatterplot(x=X[Y_means==0,0] , y=X[Y_means==0,1],color='blue')
sns.scatterplot(x=X[Y_means==1,0] , y=X[Y_means==1,1],color='red')
sns.scatterplot(x=X[Y_means==2,0] , y=X[Y_means==2,1],color='yellow')
sns.scatterplot(x=X[Y_means==3,0] , y=X[Y_means==3,1],color='green')
      
plt.xlabel("cgpa")
plt.ylabel("iq")
plt.show()
```
