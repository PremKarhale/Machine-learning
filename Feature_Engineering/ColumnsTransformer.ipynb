# Aaam Jindagi

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(df.drop(columns=['Has_Covid'] , axis=1),df['Has_Covid'],test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#adding a simple Imputer to the fever col 
si = SimpleImputer()
X_train_fever=si.fit_transform(X_train[['Fever']])
# also for the test data 
X_test_fever = si.fit_transform(X_test[['Fever']]) 
```

```python
# applying Ordinal encoding on cough col 
oe =OrdinalEncoder(categories=[['mild','strong']])
X_train_cough = oe.fit_transform(X_train[['Cough']])
# for test data 
X_test_cough = oe.fit_transform(X_test[['Cough']])

```

```python
#OneHotEncoding on gender and city
ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.float64)
X_train_gender_city = ohe.fit_transform(X_train[['Gender','City']])

# OHE for the test data 
X_test_gender_city = ohe.fit_transform(X_test[['Gender','City']])
X_train_gender_city.shape
X_test_gender_city.shape
```

```python
#Extracting age
X_train_age = X_train.drop(columns=['Gender','Fever','Cough','City'],axis=1).values
# also from test data 
X_test_age = X_test.drop(columns=['Gender','Fever','Cough','City'],axis=1).values
X_train_age.shape
```

```python
X_train_transformed = np.concatenate((X_train_age,X_train_gender_city,X_train_cough,X_train_fever),axis=1)
#for testing dataset 
X_test_transformed = np.concatenate((X_test_age,X_test_gender_city,X_test_cough,X_test_fever),axis=1)
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['Fever']),
    ('tnf2',OrdinalEncoder(categories=[['mild','strong']]),['Cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['Gender','City'])
],remainder='passthrough')
```
