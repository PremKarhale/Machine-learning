```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv(r"c:\Users\Admin\Documents\DataSets\Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df = df.drop(columns=['PassengerId','Name','Ticket','Cabin'])
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
# Applying Simple Imputer 
si_age = SimpleImputer()
si_embarked = SimpleImputer(strategy='most_frequent')

df['Age']= si_age.fit_transform(df[['Age']]).ravel()  # .ravel() returns a 1D array 
df['Embarked']=si_embarked.fit_transform(df[['Embarked']]).ravel()
```

```python
df.isnull().sum()
```

```python
# Step 1 Apply train test split 
X_train , X_test , Y_train , Y_test = train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2 ,random_state=42)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train['Sex']
```

```python
# Applying OHE on Sex or Embarked 
ohe_sex = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
# for training data 
X_train_Sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train[['Embarked']])
#for testing data 
X_test_Sex = ohe_sex.fit_transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.fit_transform(X_test[['Embarked']])
```

```python
X_train_Sex.shape # has two col male , female 
```

```python
X_train_embarked.shape # has three col C , S , Q
```

```python
X_train_rem = X_train.drop(columns=['Sex','Embarked']) #train
X_test_rem = X_test.drop(columns=['Sex','Embarked'])  # test
```

```python
X_train_rem.shape
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_Sex,X_train_embarked),axis=1)
X_test_transformed = np.concatenate((X_test_rem,X_test_Sex,X_test_embarked),axis=1)
```

```python
X_train_transformed # give a numpy array 
```

```python
X_train_transformed.shape
```

```python
X_test_transformed.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train_transformed,Y_train)
```

```python
y_pred = model.predict(X_test_transformed)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score , precision_score
```

```python
Accuracy=(np.round(accuracy_score(Y_test,y_pred),2)*100).item()
print(f'Accuracy is :{Accuracy} %')
Pricision=(np.round(precision_score(Y_test,y_pred),2)*100).item()
print(f'Precision is :{Pricision} %')
```
